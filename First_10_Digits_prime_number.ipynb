### Check if the number is prime

```python
def is_prime(n):
    n = int(n)
    if n < 2:
        return False
    div = 2
    while div < math.floor(math.sqrt(n)):
        if n % div == 0:
            return False
        div += 1
    return True

#7427466391??
```

### [Get e] 1. By Taylor Series

```python
def get_e_by_TS():
    e = Decimal(0)
    f = Decimal(1)
    n = Decimal(1)

    while True:
        olde = e
        e += Decimal(1) / f
        if e == olde: # if there was no change in the 40 places, stop.
            break
        f *= n
        n += Decimal(1)
    return e
```

### [Get e] 2. By Definition

```python
def get_e_by_Df():
    e = Decimal(0)
    n = Decimal(1)
    idx = 0
    while True:
        old_e = e
        e = (1 + 1/n)**n
        if e == old_e:
            break
        n += Decimal(1)

        if idx % 10000000 == 0:
            print(idx, e)

        idx += 1

    return e
```
